```python
N_GROUPS = 10
SEARCH_SIZE = len(suburbs) // N_GROUPS

suburb2url = lambda suburb, postcode: '+'.join(suburb.split()).lower() + ',+nsw+' + postcode
```

```python
%%time

pyautogui.PAUSE = 1
pyautogui.FAILSAFE = True

url_bar = (655, 78) # Search bar
notepad = (275, 1039)
close_notepad = (1794, 54)
close_tab = (1062, 22) # of source html page

TIME = 0.1

def ctrl(key, time=TIME):
    pyautogui.keyDown('ctrl')
    pyautogui.keyDown(key)
    sleep(time)
    pyautogui.keyUp('ctrl') 
    pyautogui.keyUp(key) 
    sleep(time)


i = 0
j = 0

# Open new tab to get started
ctrl('t')

while subs := suburbs[i: i + SEARCH_SIZE]:
    print(f'\nSUBURBS {i+1}/{len(suburbs)}')
        
    page_num = 1
    
    while True:
        
        print(page_num, end=', ')
        url = f'https://www.realestate.com.au/rent/with-1-bedroom-between-400-900-in-{"%3b+".join([suburb2url(s, p) for p, s in subs])}/list-{page_num}?maxBeds=2&includeSurrounding=false&source=refinement'    
        page_num += 1

        pyautogui.click(*url_bar)
        sleep(TIME)

        ctrl('a')

        pyautogui.typewrite(['delete'])
        sleep(TIME)
        pyautogui.typewrite(f'{url}\n')
        pyautogui.moveTo(100, 100, 8)

        ctrl('u')

        ctrl('a')

        ctrl('c')

        pyautogui.click(*notepad)
        sleep(TIME)

        ctrl('v')

        ctrl('s')

        pyautogui.typewrite(['delete'])
        sleep(TIME)

        filename = f'real_estate_{j}.txt'
        j += 1

        pyautogui.typewrite(filename)
        sleep(TIME)
        pyautogui.typewrite(['enter'])
        sleep(TIME)

        pyautogui.click(*close_notepad)
        sleep(TIME)

        pyautogui.click(*close_tab)
        sleep(TIME)

        remove = False
        with open(LOCATION + filename, encoding="utf8") as fp:
            text = fp.read()
            if 'find anything that quite matches your search.' in text or 'Card__Box' not in text:
                remove = True
                
        if remove:
            os.remove(LOCATION + filename)
            break
    
    i += SEARCH_SIZE
```
